```

import ipywidgets as widgets
from IPython.display import display, clear_output

output = widgets.Output()

question = widgets.Label("💬 Do you love me?")

button_yes = widgets.Button(description="Yes ❤️", button_style='success')
button_no = widgets.Button(description="No 💔", button_style='danger')
button_cancel = widgets.Button(description="Cancel 🤷", button_style='warning')

def on_yes_clicked(b):
    with output:
        clear_output()
        print("😊 I always knew you loved me.")

def on_no_clicked(b):
    with output:
        clear_output()
        print("😢 Please try again.")
        display_ui()

def on_cancel_clicked(b):
    with output:
        clear_output()
        print("❗ You must answer the question.")
        display_ui()

def display_ui():
    display(question, widgets.HBox([button_yes, button_no, button_cancel]))

button_yes.on_click(on_yes_clicked)
button_no.on_click(on_no_clicked)
button_cancel.on_click(on_cancel_clicked)

display_ui()
display(output)
```
