```python
# Main UI logic
import ipywidgets as widgets
from IPython.display import display, clear_output

output = widgets.Output()

question = widgets.Label("üí¨ Do you love me?")

button_yes = widgets.Button(description="Yes ‚ù§Ô∏è", button_style='success')
button_no = widgets.Button(description="No üíî", button_style='danger')
button_cancel = widgets.Button(description="Cancel ü§∑", button_style='warning')

def on_yes_clicked(b):
    with output:
        clear_output()
        print("üòä I always knew you loved me.")

def on_no_clicked(b):
    with output:
        clear_output()
        print("üò¢ Please try again.")
        display_ui()

def on_cancel_clicked(b):
    with output:
        clear_output()
        print("‚ùó You must answer the question.")
        display_ui()

def display_ui():
    display(question, widgets.HBox([button_yes, button_no, button_cancel]))

# Connect buttons to actions
button_yes.on_click(on_yes_clicked)
button_no.on_click(on_no_clicked)
button_cancel.on_click(on_cancel_clicked)

# Display UI
display_ui()
display(output)
```
